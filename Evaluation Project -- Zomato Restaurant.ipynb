# Project Description

Zomato Data Analysis is one of the most useful analysis for foodies who want to taste the best
cuisines of every part of the world which lies in their budget. This analysis is also for those who

# Data Variable

• Restaurant Id: Unique id of every restaurant across various cities of the world

• Restaurant Name: Name of the restaurant

• Country Code: Country in which restaurant is located

• City: City in which restaurant is located

• Address: Address of the restaurant

• Locality: Location in the city

• Locality Verbose: Detailed description of the locality

• Longitude: Longitude coordinate of the restaurant&#39;s location

• Latitude: Latitude coordinate of the restaurant&#39;s location

• Cuisines: Cuisines offered by the restaurant

• Average Cost for two: Cost for two people in different currencies ��

• Currency: Currency of the country

• Has Table booking: yes/no

• Has Online delivery: yes/ no

• Is delivering: yes/ no


• Switch to order menu: yes/no

• Price range: range of price of food

• Aggregate Rating: Average rating out of 5

• Rating color: depending upon the average rating color

• Rating text: text on the basis of rating of rating

• Votes: Number of ratings casted by people

```python
#plotting the distribution plot to checking the normal distribution 
plt.figure(figsize = (20,20))
pl = 1

for column in df:
    if pl<=11:
        ax = plt.subplot(3,4,pl)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize = 20)
    pl+=1
plt.show()
```

```python
#Renaming columns
df.rename(columns={'Country Code':'Country_Code','Average Cost for two':'Average_Cost_for_two','Has Table booking':'Has_Table_booking'},inplace=True)
```

```python
#Checking the outliers
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
```

```python
# Removing the outliers from City column
city = (q3.City + (1.5 * iqr.City))
city

index = np.where(df['City']>city)
index

df = df.drop(df.index[index])
df.shape

df.reset_index()
```

```python
# Removing the outliers from Country Code column
Country_Code = (q3.Country_Code + (1.5 * iqr.City))
Country_Code

index = np.where(df['Country_Code']>Country_Code)
index

df = df.drop(df.index[index])
df.shape

df.reset_index()
```

```python
# Removing the outliers from Currency column
currency = (q3.Currency + (1.5 * iqr.Currency))
currency

index = np.where(df['Currency']>currency)
index

df = df.drop(df.index[index])
df.shape

df.reset_index()
```

```python
# Removing the outliers from Average Cost for two  column
avg = (q3.Average_Cost_for_two + (1.5 * iqr.Average_Cost_for_two))
avg

index = np.where(df['Average_Cost_for_two']>avg)
index

df = df.drop(df.index[index])
df.shape

df.reset_index()
```

```python
# Removing the outliers from Has Table booking column
booking = (q3.Has_Table_booking + (1.5 * iqr.Has_Table_booking))
booking

index = np.where(df['Has_Table_booking']>booking)
index

df = df.drop(df.index[index])
df.shape

df.reset_index()
```

```python
#Splitting the columns into features or target variables
x=df.drop(columns=['Average_Cost_for_two'])
```

```python
x
```

```python
y=df['Average_Cost_for_two']
```

```python
y
```

```python
# transforming the feature into StandardScaler form
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso,RidgeCV,LassoCV
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=30)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#Applying Linear Regression for model Evaluation
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
predlr
```

```python
print(mean_absolute_error(predlr,y_test))
print(mean_squared_error(predlr,y_test))
print(np.sqrt(mean_squared_error(predlr,y_test)))
```

```python
#Applying KNeighborsRegressor to evaluate the model
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
#Model score
knn.score(x_test,y_test)
```

```python
#Predicting the values
predknn=knn.predict(x_test)
```

```python
predknn
```

```python
#Lasso CV 
lacv = LassoCV(alphas = None, max_iter = 1000,normalize = True)
lacv.fit(x_train,y_train)
```

```python
#alpha
alpha = lacv.alpha_
alpha
```

```python
#Lasso for model accuracy
las_knn = Lasso(alpha)
las_knn.fit(x_train,y_train)
```

```python
#Lasso Score 
las_knn.score(x_test,y_test)
```

```python
#RidgeCV score 
ridcv = RidgeCV(alphas = np.arange(0.001,0.1,0.01),normalize = True)
ridcv.fit(x_train,y_train)
```

```python
ridcv.alpha_
```

```python
rd = Ridge(alpha = ridcv.alpha_)
rd.fit(x_train,y_train)
```

```python
rd.score(x_test,y_test)
```

```python
#Using Another algorithm for regression
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
#Predicting the value
preddtr=dtr.predict(x_test)
preddtr
```

```python
print(mean_absolute_error(preddtr,y_test))
print(mean_squared_error(preddtr,y_test))
print(np.sqrt(mean_squared_error(preddtr,y_test)))
```
